```python
import os
import tifffile as tiff
import sys
import math
# 设定文件夹路径
directory_path = "Gauss_S1.00_NL0.30_B0.50_Jet"
sys.stdout=open('output_jet.txt','w')
# 初始化图像列表
images = []
nums = []
# 遍历文件夹中的文件
for filename in os.listdir(directory_path):
    if filename.startswith("jet") and filename.endswith(".tiff"):
        # 构建完整的文件路径
        file_path = os.path.join(directory_path, filename)
        # 读取图像
        image = tiff.imread(file_path)
        # 将图像添加到列表
        images.append(image)
        num=0
        # 遍历图像中的每个点
        for i in range(image.shape[0]):  # 对于高度
            for j in range(image.shape[1]):  # 对于宽度
                # 这里可以执行对每个点的操作，例如打印或处理
                pixel_value = image[i, j]
                if pixel_value>0:
                    num=num+1
        nums.append(num)
for i in range(len(nums)):
    print(nums[i])
for filename in os.listdir(directory_path):
    if filename.startswith("jet") and filename.endswith(".tiff"):
        # 构建完整的文件路径
        file_path = os.path.join(directory_path, filename)
        # 读取图像
        image = tiff.imread(file_path)
        # 将图像添加到列表
        images.append(image)
        
        # 遍历图像中的每个点
        for i in range(image.shape[0]):  # 对于高度
            for j in range(image.shape[1]):  # 对于宽度
                # 这里可以执行对每个点的操作，例如打印或处理
                pixel_value = image[i, j]
                if pixel_value>0:
                    pz=i+0.5-image.shape[0]/2
                    py=(56/2/math.pi)*math.cos((j+0.5)/56*2*math.pi)
                    px=(56/2/math.pi)*math.sin((j+0.5)/56*2*math.pi)
                    pr=math.sqrt(px*px+py*py+pz*pz)
                    px=px/pr
                    py=py/pr
                    pz=pz/pr
                    print(f'{px} {py} {pz} {pixel_value}')            
```
